```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/titanic.csv")
```

### Gráficos em setores

Esse gráfico consiste em representar informações na forma de setor circular. Para tanto, basta considerar que 100% dos dados equivalem a um setor circular de 360°. Dessa forma, metade desses dados equivale a um setor circular de 180°, e assim por diante.

#### Coluna Qualitativa: Sex
#### Valores: Male - Female

```python
%matplotlib inline
df['Sex'].value_counts().plot(kind='pie').legend(["Homem", "Mulher"])
```

#### Coluna Qualitativa: Embarked
#### Valores: S - C - Q

```python
df['Embarked'].value_counts().sort_index().plot(kind='pie').legend(["Cherbourg", "Queenstown","Southampton"],loc='lower right')
```

#### Coluna Qualitativa: Survived
#### Valores: 0 - 1

```python
df['Survived'].value_counts().sort_index().plot(kind='pie').legend(["Não Sobreviveu", "Sobreviveu"])
```

#### Coluna Qualitativa: Pclass
#### Valores: 1 - 2 - 3

```python
df['Pclass'].value_counts().sort_index().plot(kind='pie').legend(["1st", "2nd","3rd"],loc='lower left')
```

### Gráficos em barras para uma variável

Um gráfico de barras é um gráfico que usa barras com comprimentos proporcionais para comparar dados entre categorias. As barras podem ser plotadas verticalmente (gráfico de colunas) ou horizontalmente (gráfico de barras).

#### Coluna Qualitativa: Sex
#### Valores: Male - Female

```python
df['Sex'].value_counts().sort_index().plot(kind='bar').legend(['Sexo'])
```

```python
df['Sex'].value_counts().sort_index().plot(kind='barh').legend(['Sexo'])
```

#### Coluna Qualitativa: Embarked
#### Valores: S - C - Q

```python
df['Embarked'].value_counts().sort_index().plot(kind='bar').legend(['Embarque'])
```

```python
df['Embarked'].value_counts().sort_index().plot(kind='barh').legend(['Embarque'])
```

#### Coluna Qualitativa: Survived
#### Valores: 0 - 1

```python
df['Survived'].value_counts().sort_index().plot(kind='bar').legend(['Sobreviventes'])
```

```python
df['Survived'].value_counts().sort_index().plot(kind='barh').legend(['Sobreviventes'])
```
