##### 8. Como alterar a ordem das colunas de um dataframe? Para isso, respondam as três questões:
    (A) Em df, troque as colunas ’a’ e ’c’.

```python
df = pd.DataFrame(np.arange(20).reshape(-1, 5), columns=list('abcde'))
df
```

```python
df = df.reindex(columns=['c','b','a','d','e'])
df
```

    (B) Crie uma função genérica para trocar duas colunas, sem codificar os nomes das colunas.

```python
def troca_colunas(item, col1, col2):
    item = item.replace(col1,'-').replace(col2,'_')
    return item.replace('-',col2).replace("_",col1)

item = "abcde"
dados = input("Quais colunas deseja trocar: ")
colunas = troca_colunas(item,dados[0],dados[-1])
df = pd.DataFrame(np.arange(20).reshape(-1, 5), columns=list(colunas))
df
```

    (c) Classifique as colunas em ordem alfabética inversa, ou seja, columa ’e’ primeiro até a coluna ’a’ ultima.
